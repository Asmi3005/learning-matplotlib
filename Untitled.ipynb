```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('sample.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
dt=df['Month'].values
print(dt)
```

```python
vol_dt=df['Volume'].values
print(vol_dt)
```

```python
plt.plot(dt,vol_dt)
```

```python
plt.style.available
```

```python
plt.style.use('classic')
plt.plot(dt,vol_dt)
```

```python
plt.style.use('seaborn')
plt.plot(dt,vol_dt)
```

```python
plt.plot(dt,vol_dt,marker='.',markersize='10',color='brown')
```

```python
plt.plot(dt,vol_dt,marker='.',markersize='10')
plt.ylabel('Volume',fontsize=15)
plt.xlabel('Month',fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
```

```python
plt.plot(dt,vol_dt,marker='.',markersize='10')
plt.grid(axis='x')
```

```python
plt.plot(dt,vol_dt,marker='.',markersize='10')
plt.grid(axis='y')
```

```python
plt.plot(dt,vol_dt,label='first')
plt.legend(loc=(1.02,0))
```

```python
plt.plot(dt,vol_dt,label='first')
plt.legend(loc=(1.02,0))
plt.tight_layout()
plt.savefig('graph.png',dpi=300)
```
